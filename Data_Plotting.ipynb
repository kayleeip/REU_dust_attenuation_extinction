```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

file_path = r"C:\Users\kispz\OneDrive - Texas State University\REU MMA\Data\dirtygrid_dustony_2jun25\lmc2_bc10-wd01\cloudy\clumpy\lmc2_bc10-wd01_cloudy_clumpy-s6.32456e+14-t0.1-m0.1-a1_global_lum.table.fits"

with fits.open(file_path) as hdul:
    #hdul.info()
    data = hdul[1].data
    #print(data.columns)


    wavelength = data['wavelength']
    flux = data['flux']
    flux_input = data['flux_input']
    
    fig, ax1 = plt.subplots()
    
    line1, =   ax1.plot(wavelength, flux, color = 'blue', label = 'With Dust')
    ax1.set_xlabel('Wavelength ($\mu$m)')
    ax1.set_ylabel('Flux (erg s$^{-1}$ $\mu$m$^{-1}$)', color = 'blue')
    ax1.tick_params(axis = 'y', labelcolor = 'blue')
    ax1.set_yscale('log')
    # ax1.set_xscale('log')
    
    ax2 = ax1.twinx()
    line2, = ax2.plot(wavelength, flux_input, color = 'red', label = 'Input')
    ax2.set_ylabel('Flux_Input (erg s$^{-1}$ $\mu$m$^{-1}$)', color = 'red')
    ax2.tick_params(axis = 'y', labelcolor = 'red')
    ax2.set_yscale('log')

    lines = [line1, line2]
    labels = [line.get_label() for line in lines]

    ax1.legend(lines, labels, loc = 'upper right')
    # plt.xlim(0, 20)    
    # plt.ylim(1e42, 1e46)
    plt.title ('Spectra for LMC2 (cloudy_clumpy)')
    plt.show()
    savefig(r'C:\Users\kispz\OneDrive - Texas State University\REU MMA\Data\LMC_cloudy_clumpy plot')

  
```
