```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u

file_path = r"C:\Users\kispz\OneDrive - Texas State University\REU MMA\Data\dirtygrid_dustony_2jun25\lmc2_bc10-wd01\cloudy\clumpy\lmc2_bc10-wd01_cloudy_clumpy-s6.32456e+14-t0.1-m0.1-a1_global_lum.table.fits"

    
with fits.open(file_path) as hdul:
    # hdul.info()
    data = hdul[1].data
    # print(data.columns)

    wavelength = data['wavelength'] * u.micron
    
    flux = data['flux']
    
    flux_input = data['flux_input']
    flux_unc = data['flux_unc']
    
    flux_rt_d = data['flux_rt_d']
    flux_rt_d_unc = data['flux_rt_d_unc']
    
    flux_rt_s = data['flux_rt_s']
    flux_rt_s_unc = data['flux_rt_s_unc']
    
   
    mask = wavelength < 0.3 * u.micron
    
    wavelength_uv = wavelength[mask].value  
    
    flux = flux[mask]
    flux_input = flux_input[mask]
    flux_unc = flux_unc[mask]
    
    flux_rt_d = flux_rt_d[mask]
    flux_rt_d_unc = flux_rt_d_unc[mask]
    
    flux_rt_s = flux_rt_s[mask]
    flux_rt_s_unc = flux_rt_s_unc[mask]
    

    
    plt.figure()
    
    plt.plot(wavelength_uv, flux, color='blue', linewidth = 0.7, label='Flux')
    plt.plot(wavelength_uv, flux_input, color='red',linewidth = 0.7, label='Input')
    plt.plot(wavelength_uv, flux_unc, color='green', linewidth = 0.7, label='Unc')
    
    plt.plot(wavelength_uv, flux_rt_d, color='cyan', linewidth = 0.7, label='rt_d')
    plt.plot(wavelength_uv, flux_rt_d_unc, color='navy', linewidth = 0.7, label='rt_d_unc')
    plt.plot(wavelength_uv, flux_rt_s, color='lightblue',linewidth = 0.7, label='rt_s')
    plt.plot(wavelength_uv, flux_rt_s_unc, color='blueviolet', linewidth = 0.7, label='rt_s_unc')
    
    plt.xlabel('Wavelength ($\mu$m)')
    plt.ylabel('Flux (erg s$^{-1}$ µm$^{-1}$)')
    plt.yscale('log')
    plt.title('UV Spectrum for LMC2 (cloudy_clumpy)')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
```

```python
import os
import re
import pandas as pd

# Change this to wherever your `dirtygrid_dustony_2jun25` folder lives
root_dir = r"C:\Users\kispz\OneDrive - Texas State University\REU MMA\Data\dirtygrid_dustony_2jun25"

pattern = re.compile(
    r'(?P<grain_type>lmc2|rv31|smcbar)_'        # dust grain model
    r'(?P<ssp_model>bc\d+)-(?P<dust_model>wd\d+)_'
    r'(?P<geometry>cloudy|dusty|shell3)_'
    r'(?P<distribution>clumpy|homo)-'
    r's(?P<stellar_mass>[\deE\+\.-]+)-'
    r't(?P<age>[\d\.]+)-'
    r'm(?P<metallicity>[\d\.]+)-'
    r'(?P<sfh>a\d+)_global_lum\.table\.fits$'
)


records = []

for dirpath, _, files in os.walk(root_dir):
    for fn in files:
        if fn.endswith(".fits") and "global_lum.table" in fn:
            m = pattern.match(fn)
            if m:
                rec = m.groupdict()
                # store full path so you can always go back to the file
                rec["path"] = os.path.join(dirpath, fn)
                records.append(rec)

df = pd.DataFrame(records)

# Quick look
print(df.head())

# Save to CSV for easy viewing
df.to_csv("model_parameter_overview.csv", index=False)

```
